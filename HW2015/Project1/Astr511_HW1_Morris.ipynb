### Import the data

```python
%matplotlib inline
import numpy as np
import matplotlib
matplotlib.rcParams['font.size'] = 18
from matplotlib import pyplot as plt

datafile = 'Astr511HW1data.dat'
# Read in names of each column and the raw data
columnlabels = open(datafile, 'r').read().splitlines()[0][1:].split()
bigtable = np.genfromtxt(datafile)

# Make dictionary of the data where keys = column labels
data = {}
for label, column in zip(columnlabels, bigtable.T):
    data[label] = column
```

### Equation for metallicity:

$$ [\text{Fe/H}] = A + Bx + Cy + Dxy + Ex^2 + Fy^2 + Gx^2y + Hxy^2 + Ix^3 + Jy^3 \tag{1}$$

where $x = (u-g)$ and $y = (g-r)$.

### Equations from [Ivezic+ 2008](http://adsabs.harvard.edu/abs/2008ApJ...684..287I)

$$ \Delta M_r([\text{Fe/H}]) = 4.50 - 1.11[\text{Fe/H}] - 0.18[\text{Fe/H}]^2 \tag{A2}$$

$$ M_r(g-i, [\text{Fe/H}]) = M_r^0(g-i) + \Delta M_r([\text{Fe/H}]) \tag{A3}$$

$$ M_r^0(g-i) = -5.06 + 14.32x - 12.97x^2 + 6.127x^3 - 1.267x^4 + 0.0967x^5 \tag{A7}$$

where $x = (g-i)$

### Compute Distance

```python
# Compute approximate metallicity with ugr photometry as in Bond et al. 2010 (Eqn 1)
x = data['u'] - data['g']
y = data['g'] - data['r']
A, B, C, D, E, F, G, H, I, J = [-13.13, 14.09, 28.04, -5.51, -5.90, -58.68, 9.14, 
                                -20.61, 0.0, 58.20]
bondisapplicable = y < 0.6
metallicity = np.zeros_like(x)
metallicity[bondisapplicable] = (A + B*x + C*y + D*x*y + E*x**2 + F*y**2 + G*y*x**2 + 
                                 H*x*y**2 + I*x**3 + J*y**3)[bondisapplicable]
metallicity[np.invert(bondisapplicable)] = -0.6

# Put these metallicities into the data dictionary
data['metallicity'] = metallicity

# Solve Eqn A2 using the metallicity found with Eqn 1
data['DeltaM_r'] = 4.50 - 1.11*data['metallicity'] - 0.18*data['metallicity']**2

# Solve Eqn A7
x = data['g'] - data['i']
data['M_r^0'] = -5.06 + 14.32*x - 12.97*x**2 + 6.127*x**3 - 1.267*x**4 + 0.0967*x**5

# Solve Eqn A3
data['M_r'] = data['M_r^0'] + data['DeltaM_r']

# Calculate distance D=Z in parsecs
data['D'] = 10 * 10**((data['r'] - data['M_r'])/5)
```

# Question 1

```python
def logdensity(lower, upper, metallicitylimit=None, Nbins=50):
    '''
    Find density of stars in range of Z bins with
    lower < g - r < upper.
    '''
    x = data['g'] - data['r']
    incolorlimits = (lower < x)*(x < upper)
    if metallicitylimit is None: 
        inmetallicitylimit = np.ones_like(x).astype(bool)
    elif metallicitylimit['sign'] == '>': 
        inmetallicitylimit = data['metallicity'] > metallicitylimit['limit']
    else: 
        inmetallicitylimit = data['metallicity'] < metallicitylimit['limit']

    inalllimits = incolorlimits * inmetallicitylimit
    Omega = 314.0*np.pi/180 # solid angle in radians

    # Compute density histogram for stars in the g-r color range
    N_z, binedges = np.histogram(data['D'][inalllimits], Nbins, range=[0,6e4])
    z_bincenters_pc = [0.5*(binedges[i] + binedges[i+1]) for i in range(len(binedges) - 1)]
    z_bincenters_kpc = 1e-3 * np.array(z_bincenters_pc)
    V_z = [0.5*Omega*(binedges[i+1] - binedges[i])**2 for i in range(len(binedges) - 1)]
    rho_z = N_z / V_z

    # Compute limit in Z where fractional Poisson errors are <10%
    fractionalerr = 1./np.sqrt(N_z)
    maxreliableZ = np.max(z_bincenters_kpc[fractionalerr <= 0.10])

    return z_bincenters_kpc, rho_z, maxreliableZ
    
z_bincenters_kpc, rho_z, maxreliableZ = logdensity(0.2, 0.4)
    
fig, ax = plt.subplots(1, 2, figsize=(16,8))
ax[0].plot(z_bincenters_kpc, np.log(rho_z), ls='steps')
ax[0].set_ylabel(r'$\log \rho(Z) \, \approx \, \log \left( N(Z)/V(Z) \right)$')

ax[1].plot(z_bincenters_kpc, 1./np.sqrt(N_z), 'o')
ax[1].set_ylabel(r'Fractional uncertainty of $ N(Z) = 1/\sqrt{N}$')

for axis in ax:
    axis.set_xlabel('$Z$ [kpc]')
    axis.axvline(maxreliableZ, ls='--', color='k')
    y_annotate = axis.get_ylim()[1] - 0.05*abs(np.diff(axis.get_ylim()))
    axis.annotate('Reliable', xy=(maxreliableZ-2, y_annotate), ha='center', va='top', rotation=90)
    axis.annotate('Unreliable', xy=(maxreliableZ+3, y_annotate), ha='center', va='top', rotation=90)
    ylims = axis.get_ylim()
    axis.fill_between(np.linspace(maxreliableZ, np.max(z_bincenters_kpc), 10), 
                      ylims[0], ylims[1], color='k', alpha=0.1)
    axis.set_ylim(ylims)

plt.show()
```

Above I've computed the number of stars in 50 bins of $Z$, and computed the approximate log-density of stars with height by dividing the number of stars in a given $Z$ bin by the volume of that bin $Z$. 

To limit ourselves to the data that are most reliable, I've also plotted the fractional uncertainty in the number of stars in a bin $N(Z)$. From Poisson counting statistics, the uncertainty in a number of counts $N$ is $\sqrt{N}$, so 
$$\frac{\sigma_{N(Z)}}{N(Z)} = \frac{1}{\sqrt{N(Z)}}$$

I made the arbitrary decision that 10% fractional error is the upper limit of what I might call a "precise" measurement, and judging by that standard we can reliable measure the density out to roughly **35 kpc** using these data.

```python
# Additional color limits: 
# 0.4 < g − r < 0.6, 0.6 < g − r < 0.8, and 0.8 < g − r < 1.0
colorlimits = [(0.2, 0.4), (0.4, 0.6), (0.6, 0.8), (0.8, 1.0)]
fig, ax = plt.subplots(1, figsize=(12,8))

for limits in colorlimits: 
    z_bincenters_kpc, rho_z, maxreliableZ = logdensity(*limits)
    ax.plot(z_bincenters_kpc, np.log(rho_z), ls='steps', 
            label='${0} < g-r < {1}$'.format(*limits))
ax.set_ylabel(r'$\log \rho(Z)$')
ax.set_xlabel('$Z$ [kpc]')
ax.legend()
plt.show()
```

**Discussion**: The $0.2 < g-r < 0.4$ color bin has many more stars at higher distances than the other color bins. Though all color bins have similar densities at small heights, the density of the bluest stars drops off very quickly with height, while the red stars are relatively more evenly distributed. This is indicative of the young, blue stars being present near the disk, while the red stars are generally distributed with larger scale heights. 

We should expect larger systematic errors in the bluer bins because there are fewer stars to count so our uncertainties on the density are relatively larger. This is consistent with our understanding of the IMF: there are many more red stars than blue stars, so our measurement of the density of the red stars should be more precise than for blue stars.
