```python
import numpy as np
from astroML import datasets
import pyfits
import matplotlib.pyplot as plt
%matplotlib inline
```

## Load data using astroML (stars and quasars) and the VAGC (galaxies). Select a subsample from each so that plots are not cluttered.

```python
# Load star magnitude data using astroML

raw_star_data = datasets.fetch_sdss_S82standards(crossmatch_2mass=True) # imports data for 313859 stars
star_data = raw_star_data[::1000]
#star_data.dtype.names 
```

```python
# Load quasar magnitude data using astroML

raw_quasar_data = datasets.fetch_dr7_quasar() # imports data for ~100000 quasars
quasar_data = raw_quasar_data[::200]
#quasar_data.dtype.names
```

```python
# Load galaxy magnitude data using fits files from the VAGC

raw_sdss_galaxy_data = pyfits.open('object_sdss_imaging.fits')[1].data # imports data for 2506754 galaxies
raw_2mass_galaxy_data = pyfits.open('object_twomass.fits')[1].data 
gal_data_sdss = raw_sdss_galaxy_data[::1000]
gal_data_2mass = raw_2mass_galaxy_data[::1000]

# Reject galaxies with 2mass mags == 0
good_indices = (gal_data_2mass['J_M_K20FE'] != 0) * (gal_data_2mass['H_M_K20FE'] != 0) * \
                        (gal_data_2mass['K_M_K20FE'] != 0)

# Calculate magnitudes from given fluxes in units of nanomaggies
gal_sdss_mags = 22.5 - 2.5 * np.log10(gal_data_sdss['psfflux'][good_indices,:])
gal_2mass_J = gal_data_2mass['J_M_K20FE'][good_indices]
gal_2mass_H = gal_data_2mass['H_M_K20FE'][good_indices]
gal_2mass_K = gal_data_2mass['K_M_K20FE'][good_indices]
```

## Plot and annotate the color-color diagrams.

```python
# setup the figure and plot the data
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2,figsize=(12,10))

#r-i vs. g-r
ax1.plot(star_data['mmed_g'] - star_data['mmed_r'], star_data['mmed_r'] - star_data['mmed_i'], 'r.', label='Stars')
ax1.plot(quasar_data['mag_g'] - quasar_data['mag_r'], quasar_data['mag_r'] - quasar_data['mag_i'], 'b.', label='Quasars')
ax1.plot(gal_sdss_mags[:,1] - gal_sdss_mags[:,2], gal_sdss_mags[:,2] - gal_sdss_mags[:,3], 'm.', label='Galaxies')
ax1.set_xlabel('$g-r$', fontsize=18)
ax1.set_ylabel('$r-i$', fontsize=18)
ax1.set_ylim([-0.5, 2.2])
ax1.set_xlim([-0.5, 2.2])

#J-K vs. g-i
ax2.plot(star_data['mmed_g'] - star_data['mmed_i'], star_data['J'] - star_data['K'], 'r.', label='Stars')
ax2.plot(quasar_data['mag_g'] - quasar_data['mag_i'], quasar_data['mag_J'] - quasar_data['mag_K'], 'b.', label='Quasars')
ax2.plot(gal_sdss_mags[:,1] - gal_sdss_mags[:,3], gal_2mass_J - gal_2mass_K, 'm.', label='Galaxies')
ax2.plot([0, 0, 2, 4], [-0.2, 0.5, 1.1, 1.1], 'k-', linewidth=2)
ax2.plot([0.6, 0.6], [0.6, 2.5], 'k--', linewidth=2)
ax2.set_xlabel('$g-i$', fontsize=18)
ax2.set_ylabel('$J-K$', fontsize=18)
ax2.set_ylim([0, 2.5])
ax2.set_xlim([-1, 4])
ax2.annotate('Quasars', (-0.8, 2.3), fontsize=18)
ax2.annotate('Galaxies', (1.6, 2.0), fontsize=18)
ax2.annotate('Stars', (1.6, 0.3), fontsize=18)

#H-K vs. J-H
ax3.plot(star_data['J'] - star_data['H'], star_data['H'] - star_data['K'], 'r.', label='Stars')
ax3.plot(quasar_data['mag_J'] - quasar_data['mag_H'], quasar_data['mag_H'] - quasar_data['mag_K'], 'b.', label='Quasars')
ax3.plot(gal_2mass_J - gal_2mass_H, gal_2mass_H - gal_2mass_K, 'm.', label='Galaxies')
ax3.set_xlabel('$J-H$', fontsize=18)
ax3.set_ylabel('$H-K$', fontsize=18)
ax3.set_ylim([-0.2, 1.2])
ax3.set_xlim([-0.2, 1.2])

#i-K vs. g-i
ax4.plot(star_data['mmed_g'] - star_data['mmed_i'], star_data['mmed_i'] - star_data['K'], 'r.', label='Stars')
ax4.plot(quasar_data['mag_g'] - quasar_data['mag_i'], quasar_data['mag_i'] - quasar_data['mag_K'], 'b.', label='Quasars')
ax4.plot(gal_sdss_mags[:,1] - gal_sdss_mags[:,3], gal_sdss_mags[:,3] - gal_2mass_K, 'm.', label='Galaxies')
ax4.plot([0, 0, 4], [0, 1.5, 4.4], 'k-', linewidth=2)
ax4.plot([0.7, 0.7], [2.0, 5.0], 'k--', linewidth=2)
ax4.set_xlabel('$g-i$', fontsize=18)
ax4.set_ylabel('$J-K$', fontsize=18)
ax4.set_ylim([0, 5])
ax4.set_xlim([-1, 4])
ax4.annotate('Quasars', (-0.8, 4.5), fontsize=18)
ax4.annotate('Galaxies', (2.5, 4.5), fontsize=18)
ax4.annotate('Stars', (2.0, 1.5), fontsize=18)

plt.tight_layout()
```
