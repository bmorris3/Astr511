```python
#Imports, change path so system can find astroML

import astroML
import numpy as np
import pyfits
from matplotlib import pyplot as plt
from astroML.datasets import fetch_sdss_S82standards
from astroML.datasets import fetch_dr7_quasar
from matplotlib import collections  as mc

#Plot parameters I like
plt.rcParams['figure.figsize'] = (16,10)
plt.rcParams['font.size'] = 22
```

```python
#Get SDSS-2MASS data: data = SDSS-2MASS stars, qdata = SDSS quasar data via AstroML package
data = fetch_sdss_S82standards(crossmatch_2mass=True)
qdata = fetch_dr7_quasar()
```

```python
#Select quasars where z < 2.5
mask = (qdata['redshift'] < 2.5)
qzdata = qdata[mask]

#Remove quasars that don't have 2MASS photometry...try K band selection
mask = np.logical_and(np.logical_and(qzdata['mag_K'] != 0,qzdata['mag_J'] != 0),qzdata['mag_H'] != 0)
qzdata = qzdata[mask]

#Set subset constant for sampling
n = 200
ngal = 1000
```

```python
#Load galaxy data from http://sdss.physics.nyu.edu/vagc/
sdssGalaxy = pyfits.open('object_sdss_imaging.fits')[1].data[::ngal]
twomassGalaxy = pyfits.open('object_twomass.fits')[1].data[::ngal]

#Ensure galaxies used actually have 2MASS data...all J,H,K bands
mask = np.logical_and(np.logical_and(twomassGalaxy['J_M_K20FE'] != 0, twomassGalaxy['H_M_K20FE'] != 0),twomassGalaxy['K_M_K20FE'] != 0)

#Slice data
sdssGalaxy = sdssGalaxy[mask]
tmgal = np.zeros((len(sdssGalaxy),3))
tmgal[:,0] = twomassGalaxy['J_M_K20FE'][mask]
tmgal[:,1] = twomassGalaxy['K_M_K20FE'][mask]
tmgal[:,2] = twomassGalaxy['H_M_K20FE'][mask]

#Convert SDSS fluxes to proper magnitudes in all bands UGRIZ according to 
#http://sdss.physics.nyu.edu/vagc/flatfiles/object_sdss_imaging.fits.html
sgal = 22.5-2.5*np.log10(sdssGalaxy[:]['PSFFLUX'])
```

```python
def makeStarSubset(x1,x2,y1,y2,rc,n,data):
    """
    Function returns properly sampled subset of data
    Inputs:
    4 strings: 4 astroML-style photometric flags as strings (e.g. 'mmu_r' for r band)
    rc: red cutoff magnitude for x-axis data for plotting
    n: cadence of sampling of the form data[::n]
    data: data structure gotten from astroML fetch commands
    
    Outputs:
    x,y arrays of data for scatter plots of the form xRed, yRed, xBlue, yBlue
    """
    xOut = data[x1] - data[x2]
    yOut = data[y1] - data[y2]
    xOut = xOut[::n]
    yOut = yOut[::n]
    
    mask = (xOut > rc)
    xOutRed = xOut[mask]
    yOutRed = yOut[mask]
    
    mask = (xOut <= rc)
    return xOutRed, yOutRed, xOut[mask], yOut[mask]

def makeQuasarSubset(x1,x2,y1,y2,n,data):
    """
    Same as makeStarSubset function without the redness selection criteria
    """
    xOut = data[x1] - data[x2]
    yOut = data[y1] - data[y2]
    return xOut[::n] , yOut[::n]
```

```python
#Define int flags for easy data access
J = 0
K = 1
H = 2
U = 0
G = 1
R = 2
I = 3
Z = 4
```

```python
#Make subsets of data for plotting

#Make red/blue star subsets
g_rRed, r_iRed, g_rBlue, r_iBlue = makeStarSubset('mmu_g','mmu_r','mmu_r','mmu_i',0.7,n,data)
g_iRed, J_KRed, g_iBlue, J_KBlue = makeStarSubset('mmu_g','mmu_i','J','K',1,n,data)
J_HRed, H_KRed, J_HBlue, H_KBlue = makeStarSubset('J','H','H','K',0.55,n,data)
g_iRed4, i_KRed, g_iBlue4, i_KBlue = makeStarSubset('mmu_g','mmu_i','mmu_i','K',1,n,data)

#Make quasar subsets
qg_r, qr_i = makeQuasarSubset('mag_g','mag_r','mag_r','mag_i',n,qzdata)
qg_i, qJ_K = makeQuasarSubset('mag_g','mag_i','mag_J','mag_K',n,qzdata)
qJ_H, qH_K = makeQuasarSubset('mag_J','mag_H','mag_H','mag_K',n,qzdata)
qg_i, qi_K = makeQuasarSubset('mag_g','mag_i','mag_i','mag_K',n,qzdata)
```

```python
#Make 4subplot plot

#Create figure, ensure there's enough space
fig = plt.figure()
fig.subplots_adjust(left=0.1, right=0.9,
                    bottom=0.1, top=0.9,
                    hspace=0.4, wspace=0.4)

size = 0.2

#Plot correct diagram in appropriate subplot zone
ax1 = fig.add_subplot(221)
ax1.scatter(g_rRed,r_iRed,s=size,color='red',label='Red Stars')
ax1.scatter(g_rBlue,r_iBlue,s=size,color='blue',label='Blue Stars')
ax1.scatter(qg_r,qr_i,s=10,color='c',marker='o',label='Quasars')
ax1.scatter(sgal[:,G]-sgal[:,R],sgal[:,R]-sgal[:,I],s=size,color='green',label='Galaxies')

ax2 = fig.add_subplot(222)
ax2.scatter(g_iRed,J_KRed,s=size,color='red',label='Red Stars')
ax2.scatter(g_iBlue,J_KBlue,s=size,color='blue',label='Blue Stars')
ax2.scatter(qg_i,qJ_K,s=10,color='c',marker='o',label='Quasars')
ax2.scatter(sgal[:,G]-sgal[:,I],tmgal[:,J]-tmgal[:,K],s=size,color='green',label='Galaxies')

ax3 = fig.add_subplot(223)
ax3.scatter(J_HRed,H_KRed,s=size,color='red',label='Red Stars')
ax3.scatter(J_HBlue,H_KBlue,s=size,color='blue',label='Blue Stars')
ax3.scatter(qJ_H,qH_K,s=10,color='c',marker='o',label='Quasars')
ax3.scatter(tmgal[:,J]-tmgal[:,H],tmgal[:,H]-tmgal[:,K],s=size,color='green',label='Galaxies')

ax4 = fig.add_subplot(224)
ax4.scatter(g_iRed4,i_KRed,s=size,color='red',label='Red Stars')
ax4.scatter(g_iBlue4,i_KBlue,s=size,color='blue',label='Blue Stars')
ax4.scatter(qg_i,qi_K,s=10,color='c',marker='o',label='Quasars')
ax4.scatter(sgal[:,G]-sgal[:,I],sgal[:,I]-tmgal[:,K],s=size,color='green',label='Galaxies')

#Set subplot axis labels
ax1.set_ylabel('$r^* - i^*$')
ax1.set_xlabel('$g^* - r^*$')
ax2.set_ylabel('$J - K_s$')
ax2.set_xlabel('$g^* - i^*$')
ax3.set_ylabel('$H - K_S$')
ax3.set_xlabel('$J - H$')
ax4.set_xlabel('$g^* - i^*$')
ax4.set_ylabel('$i^* - K_s$')

#Set axis limits, add a legend
ax1.set_xlim(-0.5,2.2)
ax1.set_ylim(-0.5,2.2)
ax3.set_xlim(-0.2,1.2)
ax3.set_ylim(-0.2,1.2)
ax2.set_ylim(-0.1,2.5)
ax2.set_xlim(-1,4)
ax4.set_xlim(-1,4)
ax4.set_ylim(0,5)
ax2.legend(bbox_to_anchor=(1.1, 0), loc='upper left', borderaxespad=0.,fontsize=25,markerscale=8,scatterpoints=2)
    
#Add Line Segments
lines2 = [[(0, 0.4), (4,1.8)], [(1, 0.75), (1, 3)],[(0,0.4), (0,-1)]]
c2 = np.array(['black','black','black'])
lc2 = mc.LineCollection(lines2, colors=c2, linewidths=2,linestyles=u'dashed')
ax2.add_collection(lc2)
    
lines4 = [[(0.9,2), (0.9,5)], [(0,1.4), (4,4.4)],[(0,1.4), (0,0)]]
c4 = np.array(['black','black','black'])
lc4 = mc.LineCollection(lines4, colors=c4, linewidths=2,linestyles=u'dashed')
ax4.add_collection(lc4)
    
#Give it a title, display
plt.suptitle("SDSS-2MASS Cross Listed Objects")
#plt.savefig('FlemingFinalPlot.pdf',bbox_inches='tight')
plt.show()
```
