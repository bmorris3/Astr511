```python
import numpy as np
import matplotlib.pyplot as plt
from astroML import datasets
from astropy.io import fits
from IPython.display import display

% matplotlib inline
```

```python
# Fetch stars, quasars using AstroML
stars = datasets.fetch_sdss_S82standards(crossmatch_2mass=True)[::500]
quasars = datasets.fetch_dr7_quasar()

# Fetch galaxy data the old fashioned way
galaxies_SDSS_fits = fits.open('object_sdss_imaging.fits')
galaxies_2MASS_fits = fits.open('object_twomass.fits')
```

```python
# Manipulate data
# Stars
s_g_r = stars['mmed_g'] - stars['mmed_r']
s_g_i = stars['mmed_g'] - stars['mmed_i']
s_J_H = stars['J'] - stars['H']
s_r_i = stars['mmed_r'] - stars['mmed_i']
s_J_K = stars['J'] - stars['K']
s_H_K = stars['H'] - stars['K']
s_i_K = stars['mmed_i'] - stars['K']

# Quasars
q_n = 500
quasars25 = quasars[np.where(quasars['redshift'] < 2.5)]
valid_quasars = np.where((quasars25['mag_J'] != 0.0) & (quasars25['mag_K'] != 0.0) & (quasars25['mag_H'] != 0.0))
quasars25_2MASS = quasars25[valid_quasars][::q_n]

q_g_r = quasars25_2MASS['mag_g'] - quasars25_2MASS['mag_r']
q_g_i = quasars25_2MASS['mag_g'] - quasars25_2MASS['mag_i']
q_J_H = quasars25_2MASS['mag_J'] - quasars25_2MASS['mag_H']
q_r_i = quasars25_2MASS['mag_r'] - quasars25_2MASS['mag_i']
q_J_K = quasars25_2MASS['mag_J'] - quasars25_2MASS['mag_K']
q_H_K = quasars25_2MASS['mag_H'] - quasars25_2MASS['mag_K']
q_i_K = quasars25_2MASS['mag_i'] - quasars25_2MASS['mag_K']

# Galaxies
g_n = 500
galaxies_SDSS = galaxies_SDSS_fits[1].data
galaxies_2MASS = galaxies_2MASS_fits[1].data

valid_galaxies = np.where((galaxies_2MASS['J_M_K20FE'] != 0) 
                                            & (galaxies_2MASS['K_M_K20FE'] != 0) & (galaxies_2MASS['H_M_K20FE'] != 0))

galaxies_SDSS2MASS = galaxies_SDSS[valid_galaxies][::g_n]
galaxies_SDSS2MASS_ugriz = 22.5 - 2.5*np.log10(galaxies_SDSS2MASS['PSFFLUX'])
galaxies_SDSS2MASS_JHK = galaxies_2MASS[valid_galaxies][::g_n]

g_g_r = galaxies_SDSS2MASS_ugriz[:,1] - galaxies_SDSS2MASS_ugriz[:,2]
g_g_i = galaxies_SDSS2MASS_ugriz[:,1] - galaxies_SDSS2MASS_ugriz[:,3]
g_J_H = galaxies_SDSS2MASS_JHK['J_M_K20FE'] - galaxies_SDSS2MASS_JHK['H_M_K20FE']
g_r_i = galaxies_SDSS2MASS_ugriz[:,2] - galaxies_SDSS2MASS_ugriz[:,3]
g_J_K = galaxies_SDSS2MASS_JHK['J_M_K20FE'] - galaxies_SDSS2MASS_JHK['K_M_K20FE']
g_H_K = galaxies_SDSS2MASS_JHK['H_M_K20FE'] - galaxies_SDSS2MASS_JHK['K_M_K20FE']
g_i_K = galaxies_SDSS2MASS_ugriz[:,3] - galaxies_SDSS2MASS_JHK['K_M_K20FE']
```

```python
# Create basic figure
rows = 2
columns = 2

fig,ax = plt.subplots(rows,columns)
fig.set_tight_layout;
fig.set_size_inches(10,7)

# Set xlims,ylims,xlabels,ylabels as 2d+ arrays
xlims = np.array([[[-0.5,2.2],[-1,4]],[[-0.2,1.2],[-1,4]]])
ylims = np.array([[[-0.5,2.2],[-0.2,2.5]],[[-0.2,1.2],[0,5]]])
xlabels = np.array([['$g^{*}-r^{*}$','$g^{*}-r^{*}$'],['$J-H$','$g^{*}-i^{*}$']])
ylabels = np.array([['$r^{*}-i^{*}$','$J-K$'],['$H-K_{*}$','$i^{*}-K_{*}$']])

# Set data into arrays
s_x = np.array([[s_g_r,s_g_i],[s_J_H,s_g_i]])
q_x = np.array([[q_g_r,q_g_i],[q_J_H,q_g_i]])
g_x = np.array([[g_g_r,g_g_i],[g_J_H,g_g_i]])
s_y = np.array([[s_r_i,s_J_K],[s_H_K,s_i_K]])
q_y = np.array([[q_r_i,q_J_K],[q_H_K,q_i_K]])
g_y = np.array([[g_r_i,g_J_K],[g_H_K,g_i_K]])

# Plot data
for i in range(rows):
    for j in range(columns):
        ax[i][j].set_xlim(xlims[i][j]);
        ax[i][j].set_ylim(ylims[i][j]);
        ax[i][j].set_xlabel(xlabels[i][j]);
        ax[i][j].set_ylabel(ylabels[i][j]);
        ax[i][j].scatter(s_x[i][j],s_y[i][j],color='red',s=0.6);
        ax[i][j].scatter(q_x[i][j],q_y[i][j],color='cyan',facecolors='none',s=40);
        ax[i][j].scatter(g_x[i][j],g_y[i][j],color='blue',s=0.6);
        
# Add additional features
ax[0][1].annotate('QSOs',(-0.2,2));
ax[0][1].annotate('Stars',(2,0.3));
ax[0][1].annotate('Galaxies',(2,2.0));
ax[0][1].vlines(x=0,ymin=-0.2,ymax=0.5);
ax[0][1].vlines(x=0.7,ymin=0.75,ymax=2.5,linestyle='--');
ax[0][1].hlines(y=1.2,xmin=2,xmax=4);
ax[0][1].plot(np.linspace(0,2),np.linspace(0.5,1.2),color='black');

ax[1][1].annotate('QSOs',(-0.2,4));
ax[1][1].annotate('Stars',(2,0.8));
ax[1][1].annotate('Galaxies',(2.4,4.2));
ax[1][1].vlines(x=0,ymin=0,ymax=1.50);
ax[1][1].vlines(x=0.65,ymin=2,ymax=5,linestyle='--');
ax[1][1].plot(np.linspace(0,4),np.linspace(1.50,4.5),color='black');

# NOTE: There is no temperature (metalicity) based coloring in this plot.
```
