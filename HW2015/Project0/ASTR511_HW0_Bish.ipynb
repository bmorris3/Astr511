```python
import numpy as np
from matplotlib import pyplot as plt
import pyfits
import astroML
from astroML.datasets import fetch_sdss_S82standards
from astroML.datasets import fetch_dr7_quasar
```

```python
#Get data for stars and quasars from astroML
stars = fetch_sdss_S82standards(crossmatch_2mass=True)
qsos = fetch_dr7_quasar()
#Get data for galaxies from http://sdss.physics.nyu.edu/vagc-dr7/vagc2/ (in the files object_sdss_imaging.fits and object_twomass.fits)
sdssgals = pyfits.open('object_sdss_imaging.fits')[1].data[::10]
tmassgals = pyfits.open('object_twomass.fits')[1].data[::10]
```

```python
#Restrict quasar data set to z<2.5
zmask = (qsos['redshift']<2.5)
zqsos = qsos[zmask]
#Remove objects if they are missing data in required bands
qsomask = (zqsos['mag_g']!=0) & (zqsos['mag_r']!=0) & (zqsos['mag_i']!=0) & (zqsos['mag_J']!=0) & (zqsos['mag_H']!=0) & (zqsos['mag_K']!=0)
finqsos = zqsos[qsomask]
starmask = (stars['mmu_g']!=0) & (stars['mmu_r']!=0) & (stars['mmu_i']!=0) & (stars['J']!=0) & (stars['H']!=0) & (stars['K']!=0)
finstars = stars[starmask]
galmask = (tmassgals['J_M_K20FE']!=0) & (tmassgals['H_M_K20FE']!=0) & (tmassgals['K_M_K20FE']!=0)
sdssgals = sdssgals[galmask]
fintmgals = tmassgals[:][galmask]
```

```python
#Convert SDSS galaxy data into correct magnitude units
finsdssgals = 22.5-2.5*np.log10(sdssgals[:]['PSFFLUX'])

#For clarity, assign alphabetical names to indices of SDSS galaxy bands
g = 1
r = 2
i = 3
```

```python
#Plot results
import matplotlib.patches as mpatches
%matplotlib inline

fig = plt.figure()
fig.suptitle('SDSS-2MASS Sources', position=(0.5,1.0))

ax1 = fig.add_subplot(221)
ax1.scatter(finstars['mmu_g']-finstars['mmu_r'], finstars['mmu_r']-finstars['mmu_i'],color='red',s=0.0001,label='Stars')
ax1.scatter(finqsos['mag_g']-finqsos['mag_r'], finqsos['mag_r']-finqsos['mag_i'],color='blue',s=0.0001,label='Quasars')
ax1.scatter(finsdssgals[:,g]-finsdssgals[:,r],finsdssgals[:,r]-finsdssgals[:,i],s=0.0001,color='green',label='Galaxies')
ax1.set_xlim(-0.5,2)
ax1.set_ylim(-0.5,3)
ax1.set_xlabel('$g^* - r^*$')
ax1.set_ylabel('$r^* - i^*$')

ax2 = fig.add_subplot(222)
ax2.scatter(finstars['mmu_g']-finstars['mmu_i'], finstars['J']-finstars['K'],color='red',s=0.0001,label='Stars')
ax2.scatter(finqsos['mag_g']-finqsos['mag_i'], finqsos['mag_J']-finqsos['mag_K'],color='blue',s=0.0001,label='Quasars')
ax2.scatter(finsdssgals[:,g]-finsdssgals[:,i],fintmgals['J_M_K20FE'][:]-fintmgals['K_M_K20FE'][:],s=0.0001,color='green',label='Galaxies')
ax2.set_xlim(-1,4.5)
ax2.set_ylim(-0.5,3)
ax2.set_xlabel('$g^* - i^*$')
ax2.set_ylabel('$J - K_s$')

ax3 = fig.add_subplot(223)
ax3.scatter(finstars['J']-finstars['H'], finstars['H']-finstars['K'],color='red',s=0.0001,label='Stars')
ax3.scatter(finqsos['mag_J']-finqsos['mag_H'], finqsos['mag_H']-finqsos['mag_K'],color='blue',s=0.0001,label='Quasars')
ax3.scatter(fintmgals['J_M_K20FE'][:]-fintmgals['H_M_K20FE'][:],fintmgals['H_M_K20FE'][:]-fintmgals['K_M_K20FE'][:],s=0.0001,color='green',label='Galaxies')
ax3.set_xlim(-0.5,2)
ax3.set_ylim(-1,2)
ax3.set_xlabel('$J - H$')
ax3.set_ylabel('$H - K_s$')

ax4 = fig.add_subplot(224)
ax4.scatter(finstars['mmu_g']-finstars['mmu_i'], finstars['mmu_i']-finstars['K'],color='red',s=0.0001,label='Stars')
ax4.scatter(finqsos['mag_g']-finqsos['mag_i'], finqsos['mag_i']-finqsos['mag_K'],color='blue',s=0.0001,label='Quasars')
ax4.scatter(finsdssgals[:,g]-finsdssgals[:,i],finsdssgals[:,i]-fintmgals['K_M_K20FE'][:],s=0.0001,color='green',label='Galaxies')
ax4.set_xlim(-1,4.5)
ax4.set_ylim(-0.5,6)
ax4.set_xlabel('$g^* - i^*$')
ax4.set_ylabel('$i^* - K_s$')

ax2.legend(bbox_to_anchor=(1.1, 0), loc='upper left', borderaxespad=0.,fontsize=10,markerscale=100,scatterpoints=10)

fig.tight_layout()
```
