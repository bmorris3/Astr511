### Load stars

```python
from astroML.datasets import *
star = fetch_sdss_S82standards(crossmatch_2mass=True)[::1000]
ri = star['mmed_r'] - star['mmed_i']
gr = star['mmed_g'] - star['mmed_r']
jk = star['J'] - star['K']
gi = star['mmed_g'] - star['mmed_i']
hk = star['H'] - star['K']
jh = star['J'] - star['H']
ik = star['mmed_i'] - star['K']
```

### Load quasars

```python
quas0 = pyfits.open('dr7qso.fit')[1].data
# plotting only quasars z < 2.5 as in original plots
quas = quas0[(quas0['Z'] < 2.5) & (quas0['JMAG'] != 0) & 
             (quas0['HMAG'] != 0) & (quas0['KMAG'] != 0)][::100]
gi_q = quas['GMAG'] - quas['IMAG']
gr_q = quas['GMAG'] - quas['RMAG']
ri_q = quas['RMAG'] - quas['IMAG']
jk_q = quas['JMAG'] - quas['KMAG']
hk_q = quas['HMAG'] - quas['KMAG']
jh_q = quas['JMAG'] - quas['HMAG']
ik_q = quas['IMAG'] - quas['KMAG']
```

### Load galaxies

```python
sdss0 = pyfits.open('object_sdss_imaging.fits')[1].data[::1000]
twomass0 = pyfits.open('object_twomass.fits')[1].data[::1000]
```

```python
# only keep cross-matched data with JHK mags
keep = (twomass0['J_M_K20FE'] != 0) & (twomass0['H_M_K20FE'] != 0) & (twomass0['K_M_K20FE'] != 0) 
sdss = sdss0[keep]
twom = twomass0[keep]

# convert from 2MASS nanomaggies flux
flux = sdss['PSFFLUX']
mag = 22.5 - 2.5*np.log10(flux)
u, g, r, i, z = (mag[:,x] for x in np.arange(5))

gi_g = g - i
gr_g = g - r
ri_g = r - i
jk_g = twom['J_M_K20FE'] - twom['K_M_K20FE']
ik_g = i - twom['K_M_K20FE']
jh_g = twom['J_M_K20FE'] - twom['H_M_K20FE']
hk_g = twom['H_M_K20FE'] - twom['K_M_K20FE']
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16,12))
ax0, ax1, ax2, ax3 = axes.flat

starcol = 'maroon'
galcol = 'teal'
quascol = 'darkorange'
a=.7

ax0.scatter(gr, ri, alpha=a, c=starcol)
ax0.scatter(gr_q, ri_q, alpha=a, c=quascol)
ax0.scatter(gr_g, ri_g, alpha=a, c=galcol)
ax0.set_xlabel('g* - r*', fontsize=18)
ax0.set_ylabel('r* - i*', fontsize=18)
ax0.set_xlim([-0.5,2.5])
ax0.set_ylim([-0.5,2.5])
ax0.legend(['Stars', 'Quasars', 'Galaxies'], loc=2, frameon=False, prop={'size':20})

ax1.scatter(gi, jk, alpha=a, c=starcol)
ax1.scatter(gi_q, jk_q, alpha=a, c=quascol)
ax1.scatter(gi_g, jk_g, alpha=a, c=galcol)
ax1.set_xlabel('g* - i*', fontsize=18)
ax1.set_ylabel('J - K', fontsize=18)
ax1.set_xlim([-1,4])
ax1.set_ylim([0,2.5])

ax2.scatter(jh, hk, alpha=a, c=starcol)
ax2.scatter(jh_q, hk_q, alpha=a, c=quascol)
ax2.scatter(jh_g, hk_g, alpha=a, c=galcol)
ax2.set_xlabel('J - H', fontsize=18)
ax2.set_ylabel('H - K', fontsize=18)
ax2.set_xlim([-0.5,1.75])
ax2.set_ylim([-0.5,1.5])

ax3.scatter(gi, ik, alpha=a, c=starcol)
ax3.scatter(gi_q, ik_q, alpha=a, c=quascol)
ax3.scatter(gi_g, ik_g, alpha=a, c=galcol)
ax3.set_xlabel('i* - K', fontsize=18)
ax3.set_ylabel('g* - i*', fontsize=18)
ax3.set_xlim([-1,4])
ax3.set_ylim([0,7])

# add lines separating diff populations
ax1.plot([.9,.9], [.7,2.5], 'k--', lw=3)
ax1.plot([0,2.2], [.45, 1.1], 'k-', lw=3)
ax1.plot([-.01,-.01], [0,.45], 'k-', lw=3)
ax1.plot([2.2, 4], [1.11, 1.11], 'k-', lw=3)

ax3.plot([.85,.85], [2.2,7], 'k--', lw=3)
ax3.plot([0,4], [1.6, 4.5], 'k-', lw=3)
ax3.plot([-.01,-.01], [0,1.6], 'k-', lw=3)

# label populations
ax1.text(-.7, 2, 'QSOs', fontsize=20)
ax1.text(2, 1.9, 'Galaxies', fontsize=20)
ax1.text(1.5, .4, 'Stars', fontsize=20)

ax3.text(-.7, 4, 'QSOs', fontsize=20)
ax3.text(2, 6, 'Galaxies', fontsize=20)
ax3.text(1.4, 1.2, 'Stars', fontsize=20)
```
