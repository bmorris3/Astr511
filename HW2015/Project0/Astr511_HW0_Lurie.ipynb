# Homework 0 #
Make SDSS/2MASS color-color plots of stars, galaxies, and QSOs

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
%matplotlib inline
```

# Load the SDSS/2MASS Stellar Data #
I'm picking the stripe with 30 < b < 45, only using the first 1000 rows.
#### Note that the native 2MASS catalog is on the Vega system, but these magnitudes have been converted to the AB system. #####

```python
# read fits file
hdu = fits.open('/local/tmp/lurie/class/SDSS2MASS_fixedRv_b30_45.fits')[1].data
cut = np.arange(0,1000,1) # only use first 1000 rows
# Get g,r,i,J,H,K
gstar = hdu.field('g')[cut]
rstar = hdu.field('r')[cut]
istar = hdu.field('i')[cut]
Jstar = hdu.field('j')[cut]
Hstar = hdu.field('h')[cut]
Kstar = hdu.field('k')[cut]
```

# Load the SDSS/2MASS QSO DATA #
Take only the first 1000 rows

```python
hdu = fits.open('/local/tmp/lurie/class/dr7qso.fit')[1].data
cut = np.arange(0,1000,1)
# Get J,H,K
Jqso = hdu.field('JMAG')[cut]
Hqso = hdu.field('HMAG')[cut]
Kqso = hdu.field('KMAG')[cut]
# Get indices of rows that have nonzero J, H, and K 
good = np.where((Jqso!=0) & (Hqso!=0) & (Kqso !=0))
# Get g,r,i
gqso = hdu.field('GMAG')[cut][good]
rqso = hdu.field('RMAG')[cut][good]
iqso = hdu.field('IMAG')[cut][good]
Jqso = hdu.field('JMAG')[good]
Hqso = hdu.field('HMAG')[good]
Kqso = hdu.field('KMAG')[good]
```

# Load the SDSS/2MASS Galaxy Data #
Take only the first 1000 rows. Although the SDSS And 2MASS data are contained in separate files, the row numbers of objects are matched.

```python
# Get J,H,K
hdu = fits.open('/local/tmp/lurie/class/object_twomass.fits')[1].data
Jgal = hdu.field('J_M_K20FE')[0:1000]
Hgal = hdu.field('H_M_K20FE')[0:1000]
Kgal = hdu.field('K_M_K20FE')[0:1000]
# Incices of rows with J, H, and K 
good = np.where((Jgal!=0) & (Hgal!=0) & (Kgal !=0))
Jgal = Jgal[good]
Hgal = Hgal[good]
Kgal = Kgal[good]
# get g,r,i
hdu = fits.open('/local/tmp/lurie/class/object_sdss_imaging.fits')[1].data
mag = 22.5 -2.5*np.log10(hdu.field('PETROFLUX')[:1000,:]) # Convert flux to magnitudes
ggal = mag[:1000,1][good]
rgal = mag[:1000,2][good]
igal = mag[:1000,3][good]
```

# Some functions to make the plot prettier #
I define this function once rather than having to do it separately for each subplot.

```python
# Set font properties and label padding
# for some reason this must be called before plt.subplots
from matplotlib.ticker import MultipleLocator, FormatStrFormatter

def font(fam='serif', serif='Bitstream Vera Serif', size=14, weight='normal', 
           xpad= 10, ypad=10):
    
    plt.rcParams['font.family']     = fam
    #plt.rcParams['font.serif']      = serif
    plt.rcParams['font.size']       = size
    plt.rcParams['font.weight']     = weight
    plt.rcParams['xtick.major.pad'] = xpad
    plt.rcParams['ytick.major.pad'] = ypad

# Set tick locations, labels, length
def prettyfy(ax, xmaj=10, xmin=1, ymaj=10, ymin=1, xfmt='%d', yfmt='%d', 
          majl=7, majw=1.75, minl=4, minw=1.25, titlepad=15):
    ax.xaxis.set_major_locator(MultipleLocator(xmaj))
    ax.xaxis.set_minor_locator(MultipleLocator(xmin))
    ax.xaxis.set_major_formatter(FormatStrFormatter(xfmt))

    ax.yaxis.set_major_locator(MultipleLocator(ymaj))
    ax.yaxis.set_minor_locator(MultipleLocator(ymin))
    ax.yaxis.set_major_formatter(FormatStrFormatter(yfmt))

    ax.tick_params('both', length=majl, width=majw, which='major')
    ax.tick_params('both', length=minl, width=minw, which='minor')

    ax.titleOffsetTrans._t = (0., titlepad/72.) # x, y offset in points/72.
```

# Plot! #

```python
plt.rc('figure', figsize=(12, 10)) # set figure size
font(size =18)
fig, axarr = plt.subplots(2,2) # make subplot array
# g-i vs r-i
axarr[0,0].scatter((gstar-rstar),(rstar-istar),color='k', s=1)
axarr[0,0].scatter((gqso-rqso),(rqso-iqso),color='g', s=10)
axarr[0,0].scatter((ggal-rgal),(rgal-igal),color='b', s=1)
axarr[0,0].set_xlim(-0.3,1.8); axarr[0,0].set_ylim(-0.2,2)
axarr[0,0].set_xlabel('g - r'); axarr[0,0].set_ylabel('r - i')
axarr[0,0].text(0., 1.7, 'stars')
axarr[0,0].text(0., 1.5, 'QSOs', color='g')
axarr[0,0].text(0., 1.3, 'galaxies', color='b')
prettyfy(axarr[0,0], xmaj=0.5, xmin=0.1, xfmt='%.1f', ymaj=0.5, ymin=0.1, yfmt='%.1f')
# g-i vs J-K
axarr[0,1].scatter((gstar-istar),(Jstar-Kstar), color='k',s=2)
axarr[0,1].scatter((gqso-iqso),(Jqso-Kqso),color='g',s=10)
axarr[0,1].scatter((ggal-igal),(Jgal-Kgal),color='b',s=2)
axarr[0,1].set_xlim(-0.5,3.5); axarr[0,1].set_ylim(-1,2.5)
axarr[0,1].set_xlabel('g - i'); axarr[0,1].set_ylabel('J - K')
prettyfy(axarr[0,1], xmaj=1.0, xmin=0.2, xfmt='%.1f', ymaj=1.0, ymin=0.2, yfmt='%.1f')
# J-H vs H-K
axarr[1,0].scatter(Jstar-Hstar, Hstar-Kstar,color='k', s=2)
axarr[1,0].scatter(Jqso-Hqso, Hqso-Kqso,color='g', s=10)
axarr[1,0].scatter(Jgal-Hgal, Hgal-Kgal,color='b', s=2)
axarr[1,0].set_xlim(-0.4, 1.2); axarr[1,0].set_ylim(-1, 1.4)
axarr[1,0].set_xlabel('J - H'); axarr[1,0].set_ylabel('H - K')
prettyfy(axarr[1,0], xmaj=0.5, xmin=0.1, xfmt='%.1f', ymaj=1.0, ymin=0.2, yfmt='%.1f')
# g-i vs i-K
axarr[1,1].scatter(gstar-istar, istar-Kstar, color='k',s=2)
axarr[1,1].scatter(gqso-iqso, iqso-Kqso, color='g',s=10)
axarr[1,1].scatter(ggal-igal, igal-Kgal, color='b',s=2)
axarr[1,1].set_xlim(-0.4,3.6); axarr[1,1].set_ylim(-1,4.2)
axarr[1,1].set_xlabel('g - i'); axarr[1,1].set_ylabel('i - K')
prettyfy(axarr[1,1], xmaj=1.0, xmin=0.2, xfmt='%.1f', ymaj=1.0, ymin=0.2, yfmt='%.1f')

fig.subplots_adjust(hspace=0.3); fig.subplots_adjust(wspace=0.3)

plt.show()
```
