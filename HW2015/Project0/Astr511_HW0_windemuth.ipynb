### Goal: Fetch galaxies (VAGC), QSO (SDSS DR7 Catalog), and stars (astroML datasets) data; thin sample by 100x-1000x and stack griJHKs magnitudes to make readable plots###

```python
import numpy as np
import astropy 
import matplotlib.pyplot as plt
from astropy.io import fits as pyfits
from astroML import datasets
% matplotlib inline
```

```python
# fetch galaxies SDSS + 2MASS data
gal_sdss0 = pyfits.open('object_sdss_imaging.fits')[1].data[::1000]
gal_2mass0 = pyfits.open('object_twomass.fits')[1].data[::1000]
```

```python
# stack JHKs mags
gal_2mass = np.vstack((gal_2mass0['J_M_K20FE'],gal_2mass0['H_M_K20FE'],gal_2mass0['K_M_K20FE'])).T
# get rid of bad points / no detections
good = (gal_2mass[:, 0]!=0) * (gal_2mass[:, 1]!=0) * (gal_2mass[:,2]!=0)
gal_2mass=gal_2mass[good]
# turn nMgy into mag (doc: http://sdss.physics.nyu.edu/vagc/flatfiles/object_sdss_imaging.fits.html)
gal_sdss = 22.5 - 2.5*np.log10(gal_sdss0['PSFFLUX'][good])
# keep only gri for consistency 
gal_sdss = gal_sdss[:, 1:-1]
gal = np.hstack((gal_sdss,gal_2mass))
print gal_2mass.shape, gal_sdss.shape, gal.shape
```

```python
#fetch QSO data w/ cols redshift, g, r, i, J, H, K, selection flag
qso0 = np.genfromtxt('dr7qso.dat', skip_header=80, usecols=(3,6,8,10,22, 24, 26, 35))[::100]
#qso_dict = {'u':0, 'eu':1, 'g':2, 'eg':3, 'r':4, 'er':5, 'i':6, 'ei':7, 'z':8, 'ez':9, 'j':10, 'ej':11, 'h':12,'eh':13,'k':14,'ek':15}
```

```python
# quality control: select quasars with redshift < 2.5 and use uniform target selection flag (selected w/final algorithm)
# also, get rid of null detections (value=0)
qso = qso0[(qso0[:, 0] <2.5) * (qso0[:, -1]==1), 1:-1]
for ii in range(qso.shape[1]):
    clip = (qso[:, ii]==0.000)
    print ii, qso.shape
    qso = qso[~clip, :]
    print qso.shape
```

```python
# fetch stars data from astroML datasets
# doc: http://www.astroml.org/modules/generated/astroML.datasets.fetch_sdss_S82standards.html
stars0 = datasets.fetch_sdss_S82standards(crossmatch_2mass=True)[::1000]
```

```python
stars = np.vstack((stars0['mmed_g'], stars0['mmed_r'], stars0['mmed_i'], stars0['J'], stars0['H'], stars0['K'])).T
```

```python
# create a dictionary for band filters and indices in magnitude matrices
band = {'g':0,'r':1,'i':2,'j':3,'h':4,'k':5}
```

```python
# now make pretty plot with colors corresponding to those from lecture
import matplotlib.gridspec as gridspec
plt.rcParams.update({'font.size': 15})
plt.figure(figsize=(12, 12))
gs = gridspec.GridSpec(2,2, wspace=0.3, hspace=0.3)
colx1 = ['g', 'g', 'j', 'g']
colx2 = ['r', 'i', 'h', 'i']
coly1 = ['r', 'j', 'h' ,'i']
coly2 = ['i', 'k', 'k', 'k']
# creating xlabels for each subplot
xlabels = ['$g-r$', '$g-i$', '$J-H$', '$g-i$']
ylabels = ['$r-i$', '$J-K_s$', '$H-K_s$', '$i-K_s$']
for ii in range(4):
    # note gridspec subplot order is UL, UR, LL, LR
    ax = plt.subplot(gs[ii])
    ax.plot(stars[:, band[colx1[ii]]] - stars[:, band[colx2[ii]]], stars[:, band[coly1[ii]]] - stars[:, band[coly2[ii]]], 'r.', alpha=0.8,  label='Stars')
    ax.plot(gal[:, band[colx1[ii]]] - gal[:, band[colx2[ii]]], gal[:, band[coly1[ii]]] - gal[:, band[coly2[ii]]], 'm.', alpha=0.8, label='Galaxies')
    ax.plot(qso[:, band[colx1[ii]]] - qso[:, band[colx2[ii]]], qso[:, band[coly1[ii]]] - qso[:, band[coly2[ii]]], 'c.', alpha=0.8, label='QSO(z<2.5)')
    ax.set_xlabel(xlabels[ii])
    ax.set_ylabel(ylabels[ii])
    # draw in legend and "by-eye" boundaries amongst stars, QSO, & galaxies
    if ii==0:
        ax.legend(loc=2)
        ax.set_xlim(-0.5, 2)
        ax.set_ylim(-0.5, 2.2)
    if ii==1:
        ax.plot([0, 0], [ax.get_ylim()[0], 0.5], 'k-', linewidth=2, alpha=0.5)
        ax.plot([0, 2], [0.5, 1.1], 'k-', linewidth=2, alpha=0.6)
        ax.plot([2, 4], [1.1, 1.1], 'k-', linewidth=2, alpha=0.6)
        ax.plot([0.6, 0.6], [0.7, 2.5], 'k--', linewidth=2, alpha=0.6)
        ax.set_xlim(-1, 4)
        ax.set_ylim(-0.2, 2.5)
        ax.text(2, 2, 'Galaxies')
        ax.text(-.3, 2.25, 'QSO')
        ax.text(2, .25, 'Stars')
    if ii==3:
        ax.plot([0.0, 0.], [0, 1.45], 'k-', linewidth=2, alpha=0.6)
        ax.plot([0, 4], [1.45, 4.5], 'k-', linewidth=2, alpha=0.6)
        ax.plot([.7,.7], [2.1, 6], 'k--', linewidth=2, alpha=0.6)
        ax.set_xlim(-1, 4)
        ax.set_ylim(0.5, 6)
        ax.text(2, 5.5, 'Galaxies')
        ax.text(-.3, 4.25, 'QSO')
        ax.text(2, 1.5, 'Stars')
```
