```python
# Check the data by making an initial plot 

def plot_check(b1,b2,b3,b4,title_str, FS, xlabel, ylabel):
    plt.plot(b1 - b2, b3- b4, '.')
    N = len(b1)
    title = title_str + str(N) + ' objects'
    plt.title(title, fontsize=FS)
    plt.xlabel(xlabel, fontsize=FS)
    plt.ylabel(ylabel, fontsize=FS)
    plt.show()
    
plot_check(s_g,s_r,s_r,s_i,'SDSS-2MASS stars, ' ,15,'g - r','r - i')
```

### Try http://sdss.physics.nyu.edu/vagc/#form ...

The data is in fits format : reading   with  astropy

```python
from astropy.io import fits
fits_gal_sdss = fits.open('object_sdss_imaging.fits')
fits_gal_2mass = fits.open('object_twomass.fits')
```

```python
#Print some general info - I need to find out where is the data...

#fits_gal_sdss.info()
#fits_gal_sdss[0].header[:]
#fits_gal_sdss[1].header[:]
#cols1 = fits_gal_sdss[1].columns
#cols1.info

#cols2 = tbdata_gal_2mss.columns
#cols2.info
```

```python
import numpy as np
N = 1000  

# Pick only every N-th row 
tbdata_gal_sdss = fits_gal_sdss[1].data[::N] 
tbdata_gal_2mass = fits_gal_2mass[1].data[::N]

# Create a mask to remove those rows from 2MASS that have no 2MASS measurements 
mask = (tbdata_gal_2mass['J_M_K20FE'] != 0) * (tbdata_gal_2mass['H_M_K20FE'] != 0) * (tbdata_gal_2mass['K_M_K20FE'] != 0)


# Convert nanoMagnitudes to SDSS magnitudes , and select only those rows that fulfil the criteria 

sdss_nMg = tbdata_gal_sdss['PSFFLUX']
sgal = 22.5 - 2.5 * np.log10(sdss_nMg[mask,:])  

g_J = tbdata_gal_2mass['J_M_K20FE'][mask]    # galaxies  JHK and gri bands  
g_H = tbdata_gal_2mass['H_M_K20FE'][mask]
g_K = tbdata_gal_2mass['K_M_K20FE'][mask]

g_g = sgal[:,1]
g_r = sgal[:,2]
g_i = sgal[:,3]
```

```python
# plot the data to see how it looks
plt.xlim(0,2)
plt.ylim(-0.5,1.5)
plot_check(g_g,g_r,g_r,g_i,'SDSS-2MASS galaxies, ' ,15,'g - r','r - i')
```

# Plotting stars, galaxies and quasars from SDSS-2MASS

```python
from matplotlib.gridspec import GridSpec

plt.clf()

# Define the canvas to work on   and the  grid  
fig1 = plt.figure(figsize=(12,10))

# Make a different plotting function to not show the image ...

def add_labels(x,y, FS):
    plt.xlabel(x, fontsize=FS)
    plt.ylabel(y, fontsize=FS)    

def plot_bands(b1,b2,b3,b4,C,L):
    plt.scatter(b1 - b2, b3- b4,color=C, s=3, label=L)
    
# Upper Left Quarter  :  g-r  vs r-i 
ax1 = fig1.add_subplot(221)     #   gs [rows, columns ]
add_labels('g-r', 'r-i',15)
plt.xlim(-0.5, 1.8)
plt.ylim(-0.5,2)
plot_bands(s_g, s_r, s_r, s_i,'r', 'Stars') # stars
plot_bands(q_g, q_r, q_r, q_i, 'g', 'Quasars') # quasars
plot_bands(g_g, g_r, g_r, g_i, 'b', 'Galaxies') # galaxies 
plt.legend(loc='upper left', fontsize=15,numpoints=1)


# Upper Right Quarter : g-i  vs J-K

ax2 = fig1.add_subplot(222)
add_labels('g-i', 'J-K',15)
plt.xlim(-1,4)
plt.ylim(0,2.5)
plot_bands(s_g, s_i, s_J, s_K,'r', 'Stars') # stars
plot_bands(q_g, q_i, q_J, q_K, 'g', 'Quasars') # quasars
plot_bands(g_g, g_i, g_J, g_K, 'b', 'Galaxies') # galaxies 
plt.plot([0, 0, 2, 4], [-0.2, 0.5, 1.1, 1.1], 'k-', linewidth=2)
plt.plot([0.6, 0.6], [0.7, 2.5], 'k--', linewidth=2)
plt.annotate('QSOs', (-0.5, 2.0), fontsize=18)
plt.annotate('Galaxies', (1.5, 2), fontsize=18)
plt.annotate('Stars', (2.0, 0.5), fontsize=18)

# Lower Left Quarter : J-H vs H-K
ax3 = fig1.add_subplot(223)#  (gs[0:25,0:25]) 
add_labels('J-H', 'H-K',15)
plt.xlim(-0.5, 1.8)
plt.ylim(-0.5,2)
plot_bands(s_J, s_H, s_H, s_K,'r', 'Stars') # stars
plot_bands(q_J, q_H, q_H, q_K, 'g', 'Quasars') # quasars
plot_bands(g_J, g_H, g_H, g_K, 'b', 'Galaxies') # galaxies 

# Lower Right Quarter  : g-i  vs i-K 
ax4 = fig1.add_subplot(224)
add_labels('g-i', 'i-K',15)
plt.xlim(-0.5,4)
plt.ylim(0.5,5)
plot_bands(s_g, s_i, s_i, s_K,'r', 'Stars') # stars
plot_bands(q_g, q_i, q_i, q_K, 'g', 'Quasars') # quasars
plot_bands(g_g, g_i, g_i, g_K, 'b', 'Galaxies') # galaxies 
plt.plot([0, 0, 4], [0, 1.5, 4.4], 'k-', linewidth=2)
plt.plot([0.7, 0.7], [2.0, 5.0], 'k--', linewidth=2)
plt.annotate('QSOs', (-0.2, 4.0), fontsize=18)
plt.annotate('Galaxies', (2.5, 4.5), fontsize=18)
plt.annotate('Stars', (2.0, 1.5), fontsize=18)

# show all plots 
plt.show()
```
